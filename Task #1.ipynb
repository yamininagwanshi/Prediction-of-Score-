```python
import pandas as pd
import matplotlib.pyplot as plt
from guietta import _, Gui, Quit
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv('http://bit.ly/w-data')
print(df)
```

```python
#Independent variable
x = (df.iloc[:,0].values).reshape(-1,1)
print(x)
```

```python
#Dependent variable
y = df.iloc[:,1].values
print(y)
```

```python
plt.plot(x,y)
plt.show()
```

```python
plt.scatter(x,y)
plt.show()
```

```python
#Split Dataset

from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(x,y,test_size=0.1)
```

```python
# Model instantiated 
lr = LinearRegression()

#fit the model
lr.fit(X_train,y_train)   
```

```python
 #predict the new data
    
y_pred = lr.predict(X_test)  
print(X_test,y_pred)
```

```python
coefficient = lr.coef_
print(coefficient)
```

```python
inter_cept = lr.intercept_
print(inter_cept)
```

```python
#User interface
gui = Gui(
    ["Number of Hours : ",_, '__Hours__'],
    [_,['Predict '], _],
    ["Earned Scores : ", _, '__Scores__'],
    [_,Quit, _]
    )

def score_pred(gui):
    x = float(gui.Hours)
    y = 9.70680559 * x + 2.5758243207597005
    gui.Scores = y
gui.Predict = score_pred

gui.run()
```
